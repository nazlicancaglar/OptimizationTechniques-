```python
#Import Required Libraries
import pandas as pd
import numpy as np
import scipy.special as sps
import matplotlib.pyplot as plt
```

```python
#Seeing the dataset only 10 items
data = pd.read_csv('/content/data.csv')
```

```python
#Converting dataframe to a numpy array
data = np.array(data)
```

```python
#Reading input and output array from data
X = data[:,:3] # ":3" define first 3 column must taken (0,1,2), do not take column 3
Y = data [:,3] # "3" define that the 3th column will execute.
```

```python
#Initializing parameters

w = np.random.rand(3,)

#maximum number of iterations
max_iter = 500

#define an error vector to save all error values over all iterations.
error_all = []

#learning rate for gradient descent
eta = 0.5


```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))

z = w**max_iter * X
Y_hat = sigmoid(z)
```

```python
for iter in range(max_iter):
    Y_hat = sps.expit(np.dot(X, w))

    # Compute the error using the cost function
    e = (-1 / len(Y)) * np.sum(Y * np.log(Y_hat) + (1 - Y) * np.log(1 - Y_hat))
    error_all.append(e)
    # Gradient of the error
    grad_e = np.mean(np.multiply((Y_hat - Y), X.T), axis=1)

    w_old = w
    w = w - eta * grad_e

    print('epoch {0:d}, negative log-likelihood {1:.4f}, w={2}'.format(iter, e, w.T))
```

```python
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Error Values 0.5')
plt.plot(error_all)
plt.show()
```

```python
w = np.random.rand(3)
max_iter = 500
error_all = []
eta_values = [0.5, 0.3, 0.1, 0.05, 0.01]

for eta in eta_values:
    w = np.random.rand(3)
    error_iter = []

    for i in range(max_iter):
        Y_hat = sigmoid(np.dot(X, w))
        e = (-1 / len(Y)) * np.sum(Y * np.log(Y_hat) + (1 - Y) * np.log(1 - Y_hat))
        error_iter.append(e)

        grad_e = np.mean(np.multiply((Y_hat - Y), X.T), axis=1)
        w -= eta * grad_e

    error_all.append(error_iter)
```

```python
for i, eta in enumerate(eta_values):
    plt.plot(error_all[i], label='eta = {}'.format(eta))

plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Error Values with Different Learning Rates')
plt.legend()
plt.show()
```
